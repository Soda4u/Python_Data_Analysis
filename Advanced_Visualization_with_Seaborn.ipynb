```python
#lmplot with Seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'x':[1,3,5,7,9,11], 'y':[10,25,35,33,41,59]})

#plot using lmplot
#lmplot fits the regression line by default. This was removed by "fit_reg = False"
sns.lmplot(x='x',y='y',data=df, fit_reg = False)


plt.show()
```

```python
#importing csv data and using lmplot from seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("HR_comma_sep.csv")

sns.lmplot(x='satisfaction_level',y='last_evaluation',data=df,fit_reg=False, hue='left')

plt.show()
```

```python
#bar plots using seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'x':['P','Q','R','S','T','U'], 'y':[10,25,35,33,41,59]})

sns.barplot(x='x',y='y',data=df)

plt.show()
```

```python
#Distribution plot/Histogram with seaborn and custom csv file

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.displot(df.satisfaction_level)

plt.show()
```

```python
#Box plots with seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.boxplot(data=df[['satisfaction_level','last_evaluation']])

plt.show()
```

```python
#KDE plots with seaborn
#KDE(Kernel Density Estimation) plots the probability density estimation of a given continuous variable

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.kdeplot(df.satisfaction_level)


plt.show()
```

```python
#Violin plots with seaborn
#Violin plots is a combination of density and box plot that compares the range of a particular set of data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.violinplot(data=df[['satisfaction_level','last_evaluation']])

plt.show()
```

```python
#Count plots with seaborn
#Count plots is a special type of bar plot that shows the frequency of each categorical variable

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.countplot(x='salary',data=df, hue='left')

plt.show()
```

```python
#Joint plots

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'HR_comma_sep.csv'
df = pd.read_csv(file_name)

sns.jointplot(x='satisfaction_level',y='last_evaluation',data=df,kind="kde")

plt.show()
```

```python
#Heatmaps using seaborn

import seaborn as sns
import matplotlib.pyplot as plt


data = sns.load_dataset("iris")

cor_matrix = data.corr()

sns.heatmap(cor_matrix, annot = True, cmap='YlGnBu')

plt.show()
```
