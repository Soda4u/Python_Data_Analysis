```python
#Basic syntax for forming a simple visualization using matplotlib

import matplotlib.pyplot as plt
import numpy as np

a = np.linspace(0,20)

plt.plot(a,a+0.5,label = 'linear')

plt.show()
```

```python
#Plotting multiple datasets onto one plot using matplotlib

import matplotlib.pyplot as plt
import numpy as np

x1= []
for num in range(1,12,2):
    x1.append(num)

y1 = [10,25,35,33,41,59]
plt.plot(x1,y1,label='Series-1', color='blue')

x2= []
for num in range(0,12,2):
    x2.append(num)
y2 = [15,29,32,33,38,55]
plt.plot(x2,y2,label='Series-2',color='red')

plt.xlabel("X-label")
plt.ylabel("y-label")
plt.title("Multiple Python Line Graph")

plt.legend()

plt.show()

```

```python
#Scatterplot using matplotlib

import matplotlib.pyplot as plt

x= []
for num in range(1,12,2):
    x.append(num)

y = [10,25,35,33,41,59]

#Draw the scatter chart
plt.scatter(x,y,c='blue',marker='*',alpha=0.5)

plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("Scatter Chart Sample")

plt.show()
```

```python
#Line plot using matplotlib

import matplotlib.pyplot as plt

x= []
for num in range(1,12,2):
    x.append(num)

y = [10,25,35,33,41,59]

plt.plot(x,y)
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("Line Chart Sample")
plt.show()
```

```python
#Pie chart using matplotlib

import matplotlib.pyplot as plt

subjects = ["Mathematics","Science","Communication Skills","Computer Application"]
scores = [85,62,57,92]

plt.pie(scores,
       labels = subjects,
       colors = ['r','g','b','y'],
        startangle = 90,
        shadow = True,
        explode = (0,0.1,0,0),
        autopct = '%1.1f%%')
#Shadow = Shadow of the pies, explode = pull out a pie slice list of binary value


plt.title("Student Performance")
plt.show()
```

```python
#Bar plot using matplotlib

import matplotlib.pyplot as plt

movie_ratings = [1,2,3,4,5]
rating_counts=[21,45,72,89,42]

plt.bar(movie_ratings, rating_counts, color='blue')

plt.xlabel("Movie Ratings")
plt.ylabel("Rating Frequency")

plt.title("Movie Rating Distribution")

plt.show()
```

```python
#Histrogram using matplotlib

import matplotlib.pyplot as plt

employee_age = [21,28,32,34,35,35,37,42,47,55]

bins = [20,30,40,50,60]

plt.hist(employee_age,bins, rwidth=0.6)

plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("Employee Age Distribution")

plt.show()
```

```python
#plot using pandas

import pandas as pd
import matplotlib.pyplot as plt

#Create the dataframe
df = pd.DataFrame({
    'name':['Ajay','Malala','Abhijeet','Yming','Desilva','Lisa'],
    'age':[22,72,25,19,42,38],
    'gender':['M','F','M','M','M','F'],
    'country':['India','Pakistan','Bangladesh','China','Srilanka','UK'],
    'income':[2500,3800,3300,2100,4500,5500]
})

#Create a scatter plot of Income vs Age
df.plot(kind='scatter',x ='age', y='income', color='red',title='Age vs Income')
plt.show()

#Create a bar chart of Age vs Name
df.plot(kind='bar',x='name',y='age',color='blue')
plt.show()
```
