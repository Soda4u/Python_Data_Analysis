```python
#Understanding pandas Series
import pandas as pd
import numpy as np

dict1 = {0:'Ajay', 1:'Jay',2:'Vijay'}

#Creating a pandas Series
series_pd = pd.Series(dict1)
print(series_pd)

#Creating a series from array with numpy
arr1 = np.array([51,65,48,59,68])
series_np = pd.Series(arr1)
print(series_np)

#Creating a series with single scalar value
series_ss = pd.Series(10, index=range(5))
print(series_ss)
```

```python
import quandl

sunspots = quandl.get("SIDC/SUNSPOTS_A")

#Prints the first five rows of the sunspots data
sunspots.head()
#Prints the last five rows of the sunspots data
sunspots.tail()

#Select columns
sunspots_filtered_cols = sunspots[['Yearly Mean Total Sunspot Number','Definitive/Provisional Indicator']]

#show top 5 records of filtered columns
sunspots_filtered_cols.head()


#Select rows
sunspots_filtered_rows = sunspots["20020101": "20131231"]
sunspots_filtered_rows.head()


#Boolean Filter; similar to the WHERE clause in SQL
sunspots_filtered_bool = sunspots[sunspots['Yearly Mean Total Sunspot Number'] > 
                                  sunspots['Yearly Mean Total Sunspot Number'].mean()]
sunspots_filtered_bool
```

```python
#pandas DataFrame description methods
import quandl

df = quandl.get("SIDC/SUNSPOTS_A")

#Describe the dataset
df.describe()

#Counts the number of observations in each column. Easy to gauge missing values in column
df.count()

#Compute median of all columns
df.median()

#Compute the stdev of all columns
df.std()
```

```python
#Merging data

import pandas as pd

#Load data using read_csv()
dest = pd.read_csv("dest.csv")
dest.head()

tips = pd.read_csv("tips.csv")
tips.head()

#join dataframes using inner join
df_inner = pd.merge(dest,tips,on='EmpNr',how = 'inner')
df_inner.head()

#join dataframes using outer joijn
df_outer = pd.merge(dest,tips,on='EmpNr', how='outer')
df_outer.head()
```

```python
#PIVOT TABLES

import pandas as pd
import numpy as np

#Load in the data
purchase = pd.read_csv("purchase.csv")
print(purchase.head(10))

#summarize the dataframe using pivot table
pd.pivot_table(purchase, values='Number', index = ['Weather',], columns = ['Food'], aggfunc=np.sum)
```
